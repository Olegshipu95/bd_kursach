```python
NUMBER_OF_HUMANS = 600
NUMBER_OF_ABYSS = 5
NUMBER_OF_ABBEYS = 15
NUMBER_OF_OUTSIDERS = 40
NUMBER_OF_ELECTED = 100
NUMBER_OF_ABBEY_MEMBERS = 75
```

```python
LOCKED_HUMAN = set()
```

```python
import names
import random
from datetime import datetime, timedelta
from itertools import product

def get_random_int_excluding(excluded_values, a, b) -> int:
    while True:
        random_number = random.randint(a, b)
        if random_number not in excluded_values:
            return random_number
        
def get_unique_random_humand_id(last: int = NUMBER_OF_HUMANS) -> int:
    human_id = get_random_int_excluding(LOCKED_HUMAN, 1, last)
    LOCKED_HUMAN.add(human_id)
    return human_id

def get_human_id() -> int:
    return random.randint(1, NUMBER_OF_HUMANS)
```

```python
WEAPON_EPITHETS = ['Deadly', 'Fiery', 'Venomous', 'Thunderous', 'Infernal', 'Serrated', 'Lethal', 'Frosty', 'Mystic',
                   'Swift']
WEAPON_TYPES = ['Sword', 'Axe', 'Bow', 'Dagger', 'Staff', 'Hammer', 'Mace', 'Crossbow', 'Whip', 'Polearm']
WEAPON = [f"{ep} {tp}" for (ep, tp) in product(WEAPON_EPITHETS, WEAPON_TYPES)]
random.shuffle(WEAPON)

MAGIC_EPITHETS = ['Fire', 'Ice', 'Lightning', 'Telekinesis', 'Healing', 'Invisibility', 'Time Manipulation',
                  'Necromancy', 'Teleportation', 'Mind Control']
MAGIC_TYPE = ['Arcane', 'Mystical', 'Enchanting', 'Ethereal', 'Spectral', 'Astral', 'Celestial', 'Illusory', 'Radiant',
              'Cursed']
MAGIC = [f"{ep} {tp}" for (ep, tp) in product(MAGIC_EPITHETS, MAGIC_TYPE)]

GOOD_DEEDS = [
    ("Helping the elderly cross the street", 10, 365),
    ("Volunteering at an animal shelter", 15, 180),
    ("Participating in a charity marathon", 20, 120),
    ("Donating blood", 25, 90),
    ("Organizing a charity auction", 30, 180),
    ("Volunteering at an orphanage", 25, 365),
    ("Supporting local events", 15, 90),
    ("Assisting neighbors in cleaning the area", 10, 120),
    ("Participating in an ecological tree planting campaign", 20, 180),
    ("Assisting in organizing a workshop for children", 15, 150),
    ("Rescuing an animal in distress", 25, 365),
    ("Establishing and supporting a charitable foundation", 30, 365),
    ("Teaching free online courses", 20, 180),
    ("Assisting people with disabilities in daily tasks", 15, 120),
    ("Participating in an education improvement program", 25, 180),
    ("Raising funds for critically ill patients", 35, 240),
    ("Supporting a local sports club", 15, 120),
    ("Assisting young talents in development", 20, 150),
    ("Providing first aid in accidents", 25, 90),
    ("Supporting environmental initiatives", 20, 180),
    ("Donating books to a local library", 15, 90),
    ("Participating in a beach cleanup", 20, 120),
    ("Helping organize a blood drive", 25, 180),
    ("Supporting a community garden", 15, 150),
    ("Mentoring a high school student", 20, 120),
    ("Volunteering at a soup kitchen", 30, 180),
    ("Donating old clothes to a homeless shelter", 15, 90),
    ("Assisting in disaster relief efforts", 25, 180),
    ("Participating in a neighborhood watch program", 20, 120),
    ("Sponsoring a child`s education", 30, 365),
]

BAD_DEEDS = [
    ("Vandalizing public property", -20, 180),
    ("Stealing from a local store", -30, 90),
    ("Participating in a violent protest", -25, 120),
    ("Harassing and bullying others", -15, 90),
    ("Spreading false information online", -10, 180),
    ("Engaging in illegal drug activities", -35, 365),
    ("Assaulting someone", -40, 180),
    ("Driving under the influence", -30, 90),
    ("Committing fraud", -25, 120),
    ("Ignoring social distancing rules during a pandemic", -15, 90),
    ("Engaging in hate speech", -20, 180),
    ("Robbing a bank", -50, 365),
    ("Trespassing on private property", -15, 120),
    ("Engaging in cyberbullying", -20, 180),
    ("Breaking and entering", -30, 90),
    ("Illegally dumping waste", -25, 120),
    ("Forgery", -20, 180),
    ("Tax evasion", -30, 120),
    ("Animal cruelty", -40, 365),
    ("Engaging in insider trading", -35, 365),
    ("Engaging in cyberstalking", -25, 180),
    ("Littering in a public park", -15, 90),
    ("Spreading false rumors about a coworker", -20, 120),
    ("Vandalizing public transportation", -30, 180),
    ("Illegally downloading copyrighted content", -10, 90),
    ("Refusing to follow safety regulations at work", -20, 120),
    ("Participating in illegal street racing", -30, 180),
    ("Discriminating against someone based on gender", -25, 120),
    ("Selling counterfeit goods", -35, 180),
    ("Engaging in workplace harassment", -40, 180),
]


def generate_random_timestamp(start_date='1500-01-01', end_date='1600-12-31') -> str:
    start_datetime = datetime.strptime(start_date, '%Y-%m-%d')
    end_datetime = datetime.strptime(end_date, '%Y-%m-%d')

    random_days = random.randint(0, (end_datetime - start_datetime).days)
    random_date = start_datetime + timedelta(days=random_days)

    return random_date.strftime('%Y-%m-%d')


def insert_human(name: str = None, birthdate: str = None, status: str = None) -> str:
    if name is None:
        name = names.get_full_name()
    if birthdate is None:
        birthdate = generate_random_timestamp()
    if status is None:
        status = random.choice(['alive', 'dead', 'disappeared'])
    return f"INSERT INTO db_cs_human (name, date_of_birth, status) VALUES ('{name}', '{birthdate}', '{status}');\n"


def insert_abyss(mana: int = None) -> str:
    if mana is None:
        mana = random.randint(10000, 50000)
    return f"INSERT INTO db_cs_abbys (mana) VALUES ({mana});\n"


def insert_abbey(name: str = None):
    if name is None:
        name = names.get_first_name()
    return f"INSERT INTO db_cs_abbey (name) VALUES ('{name}');\n"


def insert_altar(human_id: int = None) -> str:
    if human_id is None:
        human_id = get_human_id()
    return f"INSERT INTO db_cs_altar (author_id) VALUES ({human_id});\n"


def insert_outsider(human_id: int = None, abbys_id: int = None) -> str:
    if human_id is None:
        human_id = get_unique_random_humand_id(400)
    if abbys_id is None:
        abbys_id = random.randint(1, 5)
    return f"INSERT INTO db_cs_outsider (human_id, abbys_id) VALUES ({human_id}, {abbys_id});\n"


def insert_elected(human_id: int = None, outsider_id: int = None) -> str:
    if human_id is None:
        human_id = get_unique_random_humand_id()
    if outsider_id is None:
        outsider_id = random.randint(1, NUMBER_OF_OUTSIDERS)
    return f"INSERT INTO db_cs_elected (human_id, outsider_id) VALUES ({human_id}, {outsider_id});\n"


def insert_rune(mana: int = None, elected_id: int = None) -> str:
    if mana is None:
        mana = random.randint(1000, 10000)
    if elected_id is None:
        elected_id = random.randint(1, NUMBER_OF_ELECTED)
    return f"INSERT INTO db_cs_rune (contained_mana, elected_id) VALUES ({mana}, {elected_id});\n"


def insert_ability() -> str:
    ans = ""
    for magic in MAGIC:
        ans += f"INSERT INTO db_cs_ability (description, cost) VALUES ('{magic}', {random.randint(5000, 8500)});\n"
    return ans


def insert_instrument() -> str:
    ans = ""
    STATUS = ['new', 'used', 'destroyed']
    for stat in STATUS:
        for weapon in WEAPON:
            ans += f"INSERT INTO db_cs_instrument (description, status) VALUES ('{weapon}', '{stat}');\n"
    return ans


def insert_abbey_member(human_id: int = None, abbey_id: int = None, rank: str = None) -> str:
    if human_id is None:
        human_id = get_unique_random_humand_id()
    if abbey_id is None:
        abbey_id = random.randint(1, NUMBER_OF_ABBEYS)
    if rank is None:
        rank = random.choice(['recruit', 'common', 'elder', 'grand'])
    return f"INSERT INTO db_cs_abbey_member (human_id, abbey_id, rank) VALUES ({human_id}, {abbey_id}, '{rank}');\n"


def insert_elected_ability(elected_id: int = None, ability_id: int = None) -> str:
    if elected_id is None:
        elected_id = random.randint(1, NUMBER_OF_ELECTED)
    if ability_id is None:
        ability_id = random.randint(1, len(MAGIC)) 
    return f"INSERT INTO db_cs_elected_ability (elected_id, ability_id) VALUES ({elected_id}, {ability_id});\n"

def insert_action() -> str:
    ans = ""
    for (des, delta, expiration) in GOOD_DEEDS:
        ans += f"INSERT INTO db_cs_action (rating_delta, expire_duration, description) VALUES ({delta}, {expiration}, \'{des}\');\n"
    for (des, delta, expiration) in BAD_DEEDS:
        ans += f"INSERT INTO db_cs_action (rating_delta, expire_duration, description) VALUES ({delta}, {expiration}, \'{des}\');\n"
    return ans

def insert_human_action(human_id: int = None, action_id: int = None, date: str = None) -> str:
    if human_id is None:
        human_id = get_human_id()
    if action_id is None:
        action_id = random.randint(1, len(GOOD_DEEDS) + len(BAD_DEEDS))
    if date is None:
        date = generate_random_timestamp('1700-01-01', '1800-12-31')

    return f"INSERT INTO db_cs_human_action (human_id, action_id, date) VALUES ({human_id}, {action_id}, '{date}');\n"

def insert_sacrifice() -> str:
    ans = ""
    for i in range(1, NUMBER_OF_OUTSIDERS + 1):
        ans += f"INSERT INTO db_cs_sacrifice (data, created_outsider, instrument_id, abbey_member) VALUES ('{generate_random_timestamp('1700-01-01', '1800-12-31')}', {i}, {random.randint(len(WEAPON), len(WEAPON) * 2)}, {random.randint(1, NUMBER_OF_ABBEY_MEMBERS)});\n"
    return ans
```

```python
FILENAME = "generated_dml.sql"

with open(FILENAME, 'w') as file:
    for i in range(NUMBER_OF_HUMANS // 3):
        file.write(insert_human(status='alive'))
    for i in range(NUMBER_OF_HUMANS // 3):
        file.write(insert_human(status='dead'))
    for i in range(NUMBER_OF_HUMANS // 3):
        file.write(insert_human(status='disappeared'))
    file.write("\n")

    for i in range(NUMBER_OF_ABYSS):
        file.write(insert_abyss())
    file.write("\n")

    for i in range(NUMBER_OF_ABBEYS):
        file.write(insert_abbey())
    file.write("\n")

    for i in range(1000):
        file.write(insert_altar())
    file.write("\n")

    for i in range(NUMBER_OF_OUTSIDERS):
        file.write(insert_outsider())
    file.write("\n")

    for i in range(NUMBER_OF_ELECTED):
        file.write(insert_elected())
    file.write("\n")

    for i in range(1000):
        file.write(insert_rune())
    file.write("\n")

    file.write(insert_ability())
    file.write("\n")

    file.write(insert_instrument())
    file.write("\n")

    for i in range(NUMBER_OF_ABBEY_MEMBERS):
        file.write(insert_abbey_member())
    file.write("\n")

    for i in range(100):
        file.write(insert_elected_ability())
    file.write("\n")

    file.write(insert_action())
    file.write("\n")

    for i in range(1800):
        file.write(insert_human_action())
    file.write("\n")

    file.write(insert_sacrifice())
    
    
    
```
