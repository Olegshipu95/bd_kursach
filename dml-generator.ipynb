```python
import names
import random
from datetime import datetime, timedelta
from itertools import product

WEAPON_EPITHETS = ['Deadly', 'Fiery', 'Venomous', 'Thunderous', 'Infernal', 'Serrated', 'Lethal', 'Frosty', 'Mystic', 'Swift']
WEAPON_TYPES = ['Sword', 'Axe', 'Bow', 'Dagger', 'Staff', 'Hammer', 'Mace', 'Crossbow', 'Whip', 'Polearm']
WEAPON = [f"{ep} {tp}" for (ep, tp) in product(WEAPON_EPITHETS, WEAPON_TYPES)]
random.shuffle(WEAPON)

MAGIC_EPITHETS = ['Fire', 'Ice', 'Lightning', 'Telekinesis', 'Healing', 'Invisibility', 'Time Manipulation', 'Necromancy', 'Teleportation', 'Mind Control']
MAGIC_TYPE = ['Arcane', 'Mystical', 'Enchanting', 'Ethereal', 'Spectral', 'Astral', 'Celestial', 'Illusory', 'Radiant', 'Cursed']
MAGIC = [f"{ep} {tp}" for (ep, tp) in product(MAGIC_EPITHETS, MAGIC_TYPE)]


def generate_random_timestamp(start_date='1970-01-01', end_date='2023-12-31') -> str:
    start_datetime = datetime.strptime(start_date, '%Y-%m-%d')
    end_datetime = datetime.strptime(end_date, '%Y-%m-%d')

    random_days = random.randint(0, (end_datetime - start_datetime).days)
    random_date = start_datetime + timedelta(days=random_days)

    return random_date.strftime('%Y-%m-%d')


def insert_human(name: str = None, birthdate: str = None, status: str = None) -> str:
    if name is None:
        name = names.get_full_name()
    
    if birthdate is None:
        birthdate = generate_random_timestamp()
        
    if status is None:
        status = random.choice(['alive', 'dead', 'disappeared'])
    
    return f"INSERT INTO db_cs_human (name, date_of_birth, status) VALUES ('{name}', '{birthdate}', '{status}');\n"


def insert_abyss(mana: int = None) -> str:
    if mana is None:
        mana = random.randint(10000, 50000 + 1)
    return f"INSERT INTO db_cs_abbys (mana) VALUES ({mana});\n"

def insert_abbey(name: str = None):
    if name is None:
        name = names.get_first_name()
    return f"INSERT INTO db_cs_abbey (name) VALUES ('{name}');\n"

def insert_altar(human_id: int = None) -> str:
    if human_id is None:
        human_id = random.randint(1, 120)
    return f"INSERT INTO db_cs_altar (author_id) VALUES ({human_id});\n"

def insert_outsider(human_id: int = None, abbys_id: int = None) -> str:
    if human_id is None:
        human_id = random.randint(1, 80)
    if abbys_id is None:
        abbys_id = random.randint(1, 15)
    return f"INSERT INTO db_cs_outsider (human_id, abbys_id) VALUES ({human_id}, {abbys_id});\n"

def insert_elected(human_id: int = None, outsider_id: int = None) -> str:
    if human_id is None:
        human_id = random.randint(1, 120 + 1)
    if outsider_id is None:
        outsider_id = random.randint(1, 10 + 1)
    return f"INSERT INTO db_cs_elected (human_id, outsider_id) VALUES ({human_id}, {outsider_id});\n"

def insert_rune(mana: int = None, elected_id: int = None) -> str:
    if mana is None:
        mana = random.randint(5000, 10000 + 1)
    if elected_id is None:
        elected_id = random.randint(1, 20 + 1)
    return f"INSERT INTO db_cs_rune (contained_mana, elected_id) VALUES ({mana}, {elected_id});\n"

def insert_ability() -> str:
    ans = ""
    for magic in MAGIC:
        ans += f"INSERT INTO db_cs_ability (description, cost) VALUES ('{magic}', {random.randint(1, 1000 + 1)});\n"
    return ans

def insert_instrument() -> str:
    ans = ""
    STATUS = ['new', 'used', 'destroyed']
    for stat in STATUS:
        for weapon in WEAPON:
            ans += f"INSERT INTO db_cs_instrument (description, status) VALUES ('{weapon}', '{stat}');\n"   
    return ans
```

```python
FILENAME = "generated_dml.sql"

with open(FILENAME, 'w') as file:
    for i in range(40):
        file.write(insert_human(status='alive'))
    for i in range(40):
        file.write(insert_human(status='dead'))
    for i in range(40):
        file.write(insert_human(status='disappeared'))
    file.write("\n")
    
    for i in range(5):
        file.write(insert_abyss())
    file.write("\n")
    
    for i in range(15):
        file.write(insert_abbey())
    file.write("\n")
    
    for i in range(25):
        file.write(insert_altar())
    file.write("\n")
    
    for i in range(10):
        file.write(insert_outsider())
    file.write("\n")
    
    for i in range(20):
        file.write(insert_elected())
    file.write("\n")
    
    for i in range(100):
        file.write(insert_rune())
    file.write("\n")
        
    file.write(insert_ability())
    file.write("\n")

    file.write(insert_instrument())
    file.write("\n")
```
